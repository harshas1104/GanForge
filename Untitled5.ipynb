```python
import pandas as pd
import numpy as np
from datetime import datetime

np.random.seed(42)  # for reproducibility

# Step 1: Create time range
start_date = datetime(2025, 1, 1)
end_date = datetime(2029, 1, 1)
time_range = pd.date_range(start=start_date, end=end_date, freq='h')[:-1]
print("Expected rows:", len(time_range))  # should be 4344 for 6 months hourly

# Step 2: Simulation setup
floors = 7

usage_records = []

# Base usage for fixtures (in liters/hour)
base_usage = {
    "urinal": (0.8, 2),         # per urinal
    "sink": (1, 3),             # per sink
    "tap": (0.5, 1.5),          # per tap
    "toilet": (6, 12),          # per toilet (flush volume per hour)
    "kitchen_sink": (2, 4),     # per kitchen sink
    "cleaning_sink": (2, 5),    # per cleaning/mop sink
}

# Step 3: Simulate data
for timestamp in time_range:
    floor_data = {"Timestamp": timestamp}

    for f in range(1, floors + 1):
        # Example fixture counts per floor (customize as needed)
        n_urinals = 2
        n_sinks = 4
        n_taps = 3
        n_toilets = 2
        n_kitchen_sinks = 1 if f == 1 else 0 # e.g., only ground floor has kitchen
        n_cleaning_sinks = 1

        urinal_usage = n_urinals * np.random.uniform(*base_usage["urinal"])
        sink_usage = n_sinks * np.random.uniform(*base_usage["sink"])
        tap_usage = n_taps * np.random.uniform(*base_usage["tap"])
        toilet_usage = n_toilets * np.random.uniform(*base_usage["toilet"])
        kitchen_sink_usage = n_kitchen_sinks * np.random.uniform(*base_usage["kitchen_sink"])
        cleaning_sink_usage = n_cleaning_sinks * np.random.uniform(*base_usage["cleaning_sink"])

        total_usage = (
            urinal_usage + sink_usage + tap_usage + toilet_usage +
            kitchen_sink_usage + cleaning_sink_usage
        )
        floor_data[f"Floor_{f}"] = total_usage

    usage_records.append(floor_data)

# Step 4: Convert to DataFrame and save
hourly_df = pd.DataFrame(usage_records)
print(hourly_df.shape)
print(hourly_df.head())

hourly_df.to_csv("esb3_4years_hourly_water_usage_realistic.csv", index=False)

# If running in Google Colab, you can use:
from google.colab import files
files.download("esb3_4years_hourly_water_usage_realistic.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define function to assign season by month
def get_season(month):
    if month in [3, 4, 5]:
        return 'summer'
    elif month in [6, 7, 8, 9]:
        return 'monsoon'
    elif month in [10, 11]:
        return 'autumn'
    else:
        return 'winter'

# Add season, day type, day name, and hour columns
hourly_df['Timestamp'] = pd.to_datetime(hourly_df['Timestamp'])
hourly_df['Season'] = hourly_df['Timestamp'].dt.month.apply(get_season)
hourly_df['DayOfWeek'] = hourly_df['Timestamp'].dt.dayofweek  # Monday=0, Sunday=6
hourly_df['DayType'] = hourly_df['DayOfWeek'].apply(lambda x: 'weekend' if x >= 5 else 'weekday')
hourly_df['DayName'] = hourly_df['Timestamp'].dt.day_name()
hourly_df['Hour'] = hourly_df['Timestamp'].dt.hour

# 1. Average water consumption by season and day type
season_daytype_avg = hourly_df.groupby(['Season', 'DayType'])['Floor_1'].mean().unstack()

# Plotting average water consumption by season and day type
plt.figure(figsize=(10, 6))
season_daytype_avg.plot(kind='bar', color=['skyblue', 'orange'])
plt.title('Average Water Consumption on First Floor by Season and Day Type')
plt.xlabel('Season')
plt.ylabel('Average Water Consumption (liters)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# 2. Average water consumption for specific hours and days
summer_monday_noon = hourly_df[
    (hourly_df['Season'] == 'summer') &
    (hourly_df['DayName'] == 'Monday') &
    (hourly_df['Hour'] == 12)
]
summer_sunday_noon = hourly_df[
    (hourly_df['Season'] == 'summer') &
    (hourly_df['DayName'] == 'Sunday') &
    (hourly_df['Hour'] == 12)
]

avg_summer_monday_noon = summer_monday_noon['Floor_1'].mean()
avg_summer_sunday_noon = summer_sunday_noon['Floor_1'].mean()

# Plotting average water consumption for specific hours and days
plt.figure(figsize=(6, 6))
plt.bar(['Summer Monday 12 PM', 'Summer Sunday 12 PM'],
        [avg_summer_monday_noon, avg_summer_sunday_noon],
        color=['green', 'blue'])
plt.title('Average Water Consumption on First Floor for Specific Hours and Days')
plt.ylabel('Average Water Consumption (liters)')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load CSV with timestamp and 7 floor readings
df = pd.read_csv('esb3_4years_hourly_water_usage_realistic.csv', parse_dates=['Timestamp'], index_col='Timestamp')

# Normalize all floor temperature values (0 to 1)
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(df)  # shape: (35064, 7)
```

```python
SEQ_LEN = 24
FORECAST_HORIZON = 1

X = []
y = []

for i in range(SEQ_LEN, len(data_scaled) - FORECAST_HORIZON):
    X.append(data_scaled[i-SEQ_LEN:i])            # shape: (24, 7)
    y.append(data_scaled[i + FORECAST_HORIZON - 1])  # shape: (7,)  next hour

X = np.array(X)
y = np.array(y)
```

```python
SEQ_LEN = 24
FORECAST_STEPS = 6

X = []
y = []

for i in range(SEQ_LEN, len(data_scaled) - FORECAST_STEPS):
    X.append(data_scaled[i-SEQ_LEN:i])       # past 24 hours
    y.append(data_scaled[i:i+FORECAST_STEPS])  # next 6 hours

X = np.array(X)            # shape: (samples, 24, 7)
y = np.array(y)            # shape: (samples, 6, 7)
```

```python
split_index = 26280 - SEQ_LEN - FORECAST_STEPS

X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(24, 7)),
    LSTM(32),
    Dense(6 * 7)  # output 6 future hours  7 floors = 42 values
])

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X_train, y_train.reshape(-1, 42), epochs=20, batch_size=32, validation_split=0.1)
```

```python
y_pred = model.predict(X_test).reshape(-1, 6, 7)
y_test = y_test.reshape(-1, 6, 7)

# To reverse scaling (optional):
y_pred_inv = scaler.inverse_transform(y_pred.reshape(-1, 7)).reshape(-1, 6, 7)
y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 7)).reshape(-1, 6, 7)
```

```python
# Plot for Single-Step Forecasting
import matplotlib.pyplot as plt

# Let's say you already have these from previous steps:
# y_test_inv = inverse scaled actual test targets
# y_pred_inv = inverse scaled predictions from LSTM

# Plot predictions for 168 hours (1 week)
n_plot = 168

plt.figure(figsize=(16, 10))
for i in range(7):  # Loop through each floor
    plt.subplot(4, 2, i + 1)
    plt.plot(y_test_inv[:n_plot, i], label='Actual', color='blue')
    plt.plot(y_pred_inv[:n_plot, i], label='Predicted', color='red', linestyle='--')
    plt.title(f'Floor {i + 1}')
    plt.xlabel('Hour')
    plt.ylabel('Temperature')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Plot Multi-step Prediction for One Time Slice
# Pick one sample from the test set to visualize
sample_idx = 0  # Can try other values too

plt.figure(figsize=(12, 6))

for step in range(6):  # Forecast steps: 1-hour ahead to 6-hours ahead
    plt.plot(y_test_inv[sample_idx, step], label=f'Actual Step {step+1}', linestyle='-')
    plt.plot(y_pred_inv[sample_idx, step], label=f'Predicted Step {step+1}', linestyle='--')

plt.title("Multi-step Forecasting for All 7 Floors (1 Sample)")
plt.xlabel("Floor Index (17)")
plt.ylabel("Temperature")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Plot All 6 Steps for a Specific Floor
floor = 0  # Floor_1

plt.figure(figsize=(10, 6))
plt.plot(y_test_inv[sample_idx, :, floor], label='Actual', marker='o')
plt.plot(y_pred_inv[sample_idx, :, floor], label='Predicted', marker='x')
plt.title(f"Floor 1  6-Hour Forecast")
plt.xlabel("Forecast Step (1 to 6 Hours Ahead)")
plt.ylabel("Temperature")
plt.legend()
plt.grid(True)
plt.show()
```
